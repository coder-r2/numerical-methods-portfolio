```python
import numpy as np
import matplotlib.pyplot as plt
from core import generate_mandelbrot, generate_newton_fractal, simulate_logistic_map, calculate_lyapunov_exponent
```

# 1. Visualise Mandelbrot Set

```python
print("Generating Mandelbrot Set...")
mandelbrot_data = generate_mandelbrot()

plt.figure(figsize=(8, 6))
plt.imshow(mandelbrot_data, extent=[-2.5, 1.0, -1.5, 1.5], cmap='hot', origin='lower')
plt.colorbar(label='Iterations to Divergence')
plt.title('Mandelbrot Set')
plt.xlabel('Real Axis')
plt.ylabel('Imaginary Axis')
plt.savefig('imgs/mandelbrot.png', dpi=300)
plt.show()
```

# 2. Visualize Newton's Fractal

```python
print("Generating Newton's Fractal...")
newton_data = generate_newton_fractal()

plt.figure(figsize=(8, 6))
plt.imshow(newton_data, extent=[-2.0, 2.0, -2.0, 2.0], cmap='viridis', origin='lower')
plt.colorbar(label='Converged Root Index (0, 1, 2)')
plt.title("Newton's Fractal for f(z) = z³ + 1")
plt.xlabel('Real Axis')
plt.ylabel('Imaginary Axis')
plt.savefig('imgs/newton_fractal.png', dpi=300)
plt.show()
```

# 3. Visualize Logistic Map Bifurcation

```python
print("Generating Bifurcation Diagram...")
A_vals, x_vals = simulate_logistic_map()

plt.figure(figsize=(10, 6))
plt.scatter(A_vals, x_vals, s=0.1, color='black', alpha=0.5)
plt.title('Logistic Map Bifurcation Diagram')
plt.xlabel('Control Parameter A')
plt.ylabel('Steady-state Population (x)')
plt.xlim(0.89, 4.0)
plt.ylim(0, 1)
plt.savefig('imgs/bifurcation.png', dpi=300)
plt.show()
```

# 4. Visualize Lyapunov Exponent

```python
print("Calculating Lyapunov Exponent...")
lyapunov_data = calculate_lyapunov_exponent(A=4.0, x0=0.4, n_iter=1000)

plt.figure(figsize=(10, 4))
plt.plot(range(1, 1001), lyapunov_data, color='blue', linewidth=1)
plt.axhline(y=np.log(2), color='red', linestyle='--', label='Theoretical limit: ln(2)')
plt.title('Finite-Time Lyapunov Exponent (A=4, x0=0.4)')
plt.xlabel('Iterations (n)')
plt.ylabel('Lyapunov Exponent (λ)')
plt.legend()
plt.grid(True)
plt.savefig('imgs/lyapunov.png', dpi=300)
plt.show()
```
